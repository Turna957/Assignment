```python
# curve

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

mu, sigma = 0, 1
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y, c='black')
plt.show()
```

```python
# fibonacci series
def findFibonacci(n):

    #base case n = 0 , n = 1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return findFibonacci(n-2) + findFibonacci(n-1)


def fibonacciNumbers(n):
    ans = []
    for i in range(n):
        ans.append(findFibonacci(i))
    return ans


if __name__ == '__main__':
    n = 7
    res = fibonacciNumbers(n)
    for num in res:
        print(num, end=' ')
```

```python
# Hyperbola plot
import numpy as np
import matplotlib.pyplot as plt

a = 1
b = 2

x_values = np.linspace(-5, 5, 400)

y_positive = np.sqrt((x_values**2 / a**2 - 1) * b**2)
y_negative = -y_positive

plt.plot(x_values, y_positive, label='Hyperbola: y = sqrt((x^2/a^2 - 1) * b^2)', color='b')
plt.plot(x_values, y_negative, color='b')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperbola Plot')

plt.legend()
plt.show()
```

```python
#Elliptic Curve Plot

import numpy as np
import matplotlib.pyplot as plt


a = -1
b = 1


x_curve = np.linspace(-2, 2, 400)

y_square = x_curve**3 + a * x_curve + b

y_curve_positive = np.sqrt(y_square)
y_curve_negative = -y_curve_positive


plt.plot(x_curve, y_curve_positive, label='Elliptic Curve: $y^2 = x^3 + ax + b$', color='r')

plt.plot(x_curve, y_curve_negative, color='r')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Elliptic Curve Plot')

plt.legend()

plt.show()
```
